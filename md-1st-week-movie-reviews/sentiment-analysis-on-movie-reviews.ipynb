# Clean Text

```python
# phrase = "ca n't recommend it"

def clean_text(phrase):
    phrase =  phrase.replace("doesn't", "does not ")
    phrase =  phrase.replace("ca n't", "can not ")
    phrase =  phrase.replace("n't", " not ")
    
    return phrase

train["Phrase"] = train["Phrase"].apply(clean_text)
train[["Phrase", "Phrase(origin)"]].head()
```

```python
test["Phrase"] = test["Phrase"].apply(clean_text)
test[["Phrase", "Phrase(origin)"]].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# binary = True/False
# lowercase = True/False
# ngram_range(1, 2) 
# stop_word=None a나 the가 지금은 빠져있는 데 넣을 수도 있음
# 단어를 천개만 넣었다가 만개로 늘림 stunning이 나와서

vectorizer = CountVectorizer(max_features=10000, ngram_range=(1,2))
vectorizer
```

```python
vectorizer.fit(train["Phrase"])
```

```python
X_train = vectorizer.transform(train["Phrase"])
X_train
```

* 특수기호를 처리해야 점수가 올라간다. 이에 대한 노가다가 필요하다.
* 특수기호가 몇 개 이상있으면 긍정인지 부정인지
* 띄어쓰기 + 특수기호 제거로 count_vertorize

```python
columns = vectorizer.get_feature_names()
pd.DataFrame(X_train[:100].toarray(), columns=columns).head()
```

```python
X_test = vectorizer.transform(train["Phrase"])
X_test
```

```python
y_train = train["Sentiment"]

print(y_train.shape)
y_train.head()
```

# Score
* 데이터의 양이 많아 RF의 속도가 굉장히 느리다.
* 로우 10만개 컬럼 천개, 엄청 오래 걸리고 점수도 좋지 않다.
* 보통은 RF가 잘나오지만 무비리뷰에 잘나오는 리니어 모델을 사용


```python
from sklearn.linear_model import SGDClassifier

model = SGDClassifier(random_state=37)
model
```

```python
# # from sklearn.cross_validation import cross_val_score # 아나콘다 환경 일 때
# from sklearn.model_selection import cross_val_score

# score = cross_val_score(model, X_train, y_train, cv=5).mean()

# print("Score = {0:.5f}".format(score))
# # 오늘 목표는 0.60까지 
# Score = 0.55597
```

```python
# from sklearn.cross_validation import cross_val_score
from sklearn.model_selection import cross_val_predict

y_predict = cross_val_predict(model, X_train, y_train, cv=5)

print(y_predict.shape)
y_predict[0:10]
```

```python
# 시간 될 때 metrics 자세히 볼 것
from sklearn.metrics import accuracy_score

score = accuracy_score(y_train, y_predict)
print("Score = {0:.5f}".format(score))
```

```python
import numpy as np

result = train.copy()
result["Sentiment(predict)"] = y_predict
result["Difference(Phrase)"] = np.abs(y_train - y_predict)

# 많이 틀린것으로 정렬
# result.sort_values()

print(result.shape)
result.head()
# Difference 가 0인 것은 고칠 필요가 없다.
# Difference 가 높은 것부터 고쳐나가는게 더 도움이 된다.
# train 데이터 15만개를 모두 볼 수가 없다. 그래서 고치면 많이 올라가는 것을 고친다.

# 묶여있는 센텐스ID 전체의 차이를 구한다. 하나의 센텐스에 여러개의 Phase가 있기 때문에 차이가 높은 순으로 정렬한다.
```

```python
# 센텐스 마다의 평점
# 쪼갠 것도 점수에 대한 평점을 기록함 미케니컬 텅크를 사용해서
sentiment = result.groupby("SentenceId")["Difference(Phrase)"].mean()

print(sentiment.shape)

sentiment.head()
```

```python
def find_sentiment(sentence_id):
    return sentiment.loc[sentence_id]

result["Difference(Sentence)"] = result["SentenceId"].apply(find_sentiment)
result = result.sort_values(by="Difference(Sentence)", ascending=False)

print(result.shape)
result.head()
```
