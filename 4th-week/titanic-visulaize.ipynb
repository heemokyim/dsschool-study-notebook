```python
# Distplot
# 값 하나만의 분포를 알고 싶을 때 사용한다.
# 시각화는 직관적이고 영감을 줄 수있어야 한다. 소수의 특이한 데이터가 이 분포를 방해할 때가 있다.
# 운임요금 제일 많이 낸 사람이 500달러다, 참고로 평균이 10~20달러이다.
# 전반적인 분포를 보는 것이기 때문에 500달러를 그리기 위해 다른 아이들을 밀어버렸다.
# 시각화 할 때 구겨져 보인다. 할 때 아웃라이어를 제거하고 본다.
```

```python
sns.distplot(train["Fare"])
```

```python
sns.distplot(train["Fare"], hist=False)
```

```python
# 100달러 이하의 데이터만 본다. 10달러에 많이 분포 되어 있으며, 20, 40, 60 달러에 많은 분포가 있다.
low_fare = train[train["Fare"] < 100]
sns.distplot(low_fare["Fare"], hist=False)
```

```python
low_fare = train[train["Fare"] < 100]

perish = low_fare[low_fare["Survived"] == 0 ]
survived = low_fare[low_fare["Survived"] == 1]

sns.distplot(perish["Fare"], hist=False, label="Perish")
sns.distplot(survived["Fare"], hist=False, label="Survived")
```

```python
# 노말 디스트리뷰션, 가우시안분포 정규분포가 안 나오면 값이 이상하다.
# 이렇게 한쪽으로 치우친 그래프를 어떻게 정규화 시키는지 고민
```

```python
# Lmplot
```

```python
# 가운데 그어진 선은 회귀선으로 분포가 어느 방향으로 가는지를 보여준다. 여기에서는 큰 의미가 없기 때문에 fit_reg옵션으로 지워준다.
# 나이와 운임요금이 어느정도일 때 생존율이 높아질까?

sns.lmplot(data=train, x="Age", y="Fare", fit_reg=False, hue="Survived")
```

```python
# 낮은 운임요금에 어느정도의 분포가 있는지
# 살아남으려면 돈이 많거나 7살 미만으로 아주 어리다는 가설을 세워 볼 수도 있을거 같다.
low_fare = train[train["Fare"] < 100]

sns.lmplot(data=low_fare, x="Age", y="Fare", fit_reg=False, hue="Survived")
```
