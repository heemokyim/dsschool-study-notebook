## 컬럼 추가

```python
train["DatasetType"] = "Titanic"
# train
```

```python
# row 갯수가 동일해야 함
# train["id"] = list(range(0,891))
train["Nationality_FR"] = train["Embarked"] == "C"
```

```python
train["Nationality_EN"] = (train["Embarked"] == "S") | (train["Embarked"] == "C" ) 
```

```python
train["FamilySize"] = (train["SibSp"] + train["Parch"]) + 1
# train[["FamilySize", "SibSp", "Parch"]]
```

```python
# 나이가 빈값에는 평균이 들어가도록 함
mean_age = train["Age"].mean()

# train.loc[train["Age"].isnull(), "Age"] = train["Age"].mean()
# mean_age 변수 사용 예제로 변경
train.loc[train["Age"].isnull(), "Age"] = mean_age

# 컬럼을 지정해 주지 않으면 모든 컬럼을 변경한다.
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1
```

```python
# 티켓 요금에 따라 Low, Medium, High로 분류
train["Fare_Low"] = train["Fare"] < 10
train["Fare_Medium"] = (train["Fare"] >= 10) & (train["Fare"] < 50 )
train["Fare_High"] = train["Fare"] >= 50

# train[["Fare", "Fare_Low", "Fare_Medium", "Fare_High"]]
```

```python
# csv로 생성
train.to_csv("data/modified_train.csv")
```
