```python
import pandas as pd
```

# Load Dataset

```python
train = pd.read_csv("data/train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("data/test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1

print(train.shape)
train.head()
```

```python
test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1

print(test.shape)
test.head()
```

```python
### Fill in missing fares
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = train["Fare"].mean()
test[test["Fare"].isnull()]
```

```python
### Encode Embarked
# C, S, Q 3개가 있음 0, 1, 2 로 바꿔주면 디시전트리가 2xS=Q? 이런 오해 방지
# C => 1 0 0
# S => 0 1 0 
# Q => 0 0 1
```

```python
# 0 == False
# 1 == True
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
# Family Size
train["FamilySize"] = (train["SibSp"] + train["Parch"]) + 1

print(train.shape)

test["FamilySize"] = (test["SibSp"] + test["Parch"]) + 1

print(test.shape)
test[["FamilySize"]].head()
```

```python
# Band Age

train

# Train
```

```python
# scikit-learn
# 레이지로딩 필요한 것만 불러옴
# 피처와 레이블(Survived) 
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=5)
model
```

```python
feature_names = ["Pclass", "Sex", "Fare", "FamilySize", "Embarked_C", "Embarked_S", "Embarked_Q"]
train_features = train[feature_names]

print(train_features.shape)
train_features.head()
```

```python
test_features = test[feature_names]

print(test_features.shape)
test_features.head()
```

```python
label_name = "Survived"

train_label = train[label_name]

print(train_label.shape)
train_label.head()
```

```python
# 학습시킴
model.fit(train_features, train_label)
```

```python
predictions = model.predict(test_features)
print(predictions.shape)
predictions[0:10]
```
